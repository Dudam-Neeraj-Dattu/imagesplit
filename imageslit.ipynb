# Program to split an image and store as patches in Julia

Here we are using jupyter notebook with Julia as its Kernel

As part of first step, we need to import Pkg library and add Images.jl to the program, then we can use this to load the actual image

```julia
# import Pkg
# Pkg.add("Images")
using Images
```

Loading the image to a variable using load function 

```julia
img = load("./lh.jpg")
```

```julia
println(typeof(img)) #to get to know type of the image
size(img) #to know the dimensions of the image
```

Now We need to split the image into patches and store each patch in a list

```julia
patch_size = 100

# Get the dimensions of the image
sie = size(img)

# Initialize an empty list to store patches
patches = []

# Iterate over rows and columns to extract patches
for y in 1:patch_size:sie[1]
    for x in 1:patch_size:sie[2]
        # Extract the patch
        patch = img[y:min(y+patch_size-1, sie[1]), x:min(x+patch_size-1, sie[2])]
        push!(patches, patch)
    end
end
```

```julia
println(size(patches)) #size of the list and its type
print(typeof(patches))
```

```julia
size(patches[1]) #to know the size of each patch and its type.
typeof(patches[1])
```

Now the image is being splitted and now we want to use a library called MosaicViews to view the image as collection of patches, so import the library and use it.

```julia
#import Pkg; Pkg.add("MosaicViews")
using MosaicViews
```

Writing a program to get a string of the form "patches[i]" such that this can be given as input to the mosaicview function

```julia
s = ""
for i in 1:size(patches)[1]
    s = s * "patches[$i], "
end
print(s)
```
